# Data Understanding Notebook

Load and explore the crop price dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Try to import seaborn (optional)
try:
    import seaborn as sns
    sns.set_style('whitegrid')
    SEABORN_AVAILABLE = True
except ImportError:
    SEABORN_AVAILABLE = False
    print("Note: seaborn not installed. Using matplotlib only.")

plt.rcParams['figure.figsize'] = (12, 6)
print("Libraries loaded successfully!")
```

## 1. Load Dataset

```python
# Load CSV file from raw_data folder
df = pd.read_csv('../dataset/raw_data/crop_prices.csv')
print('Data loaded successfully!')
print(f'Shape: {df.shape}')
print(f'File size: {df.memory_usage(deep=True).sum() / 1024:.2f} KB')
```

## 2. Dataset Information

```python
# Display columns and their data types
print('Columns and Data Types:')
print(df.dtypes)
print(f'\nTotal Rows: {len(df)}')
print(f'Total Columns: {len(df.columns)}')
print(f'\nColumn Names: {df.columns.tolist()}')
```

## 3. First 10 Rows

```python
# Display first 10 rows
print('First 10 rows of data:')
df.head(10)
```

## 4. Check for Missing Values

```python
# Check missing values in each column
print('Missing Values Count:')
print(df.isnull().sum())
print(f'\nPercentage of Missing Values:')
print((df.isnull().sum() / len(df) * 100).round(2))
print(f'\nTotal Missing Values: {df.isnull().sum().sum()}')
```

## 5. Check for Duplicates

```python
# Check for duplicate rows
print(f'Total Duplicate Rows: {df.duplicated().sum()}')
print(f'Percentage of Duplicates: {(df.duplicated().sum() / len(df) * 100).round(2)}%')

# Show duplicate rows if any
if df.duplicated().sum() > 0:
    print('\nDuplicate Rows:')
    print(df[df.duplicated(keep=False)])
```

## 6. Statistical Summary

```python
# Display descriptive statistics
print('Statistical Summary:')
df.describe()
```

## 7. Key Data Insights

```python
# Extract key insights from the data
print('\n=== KEY INSIGHTS ===')
print(f'\nUnique Commodities: {df["Commodity"].nunique()}')
print(f'Commodity List: {df["Commodity"].unique().tolist()}')
print(f'\nPrice Statistics:')
print(f'  Minimum Price: ${df["Modal Price"].min():.2f}')
print(f'  Maximum Price: ${df["Modal Price"].max():.2f}')
print(f'  Average Price: ${df["Modal Price"].mean():.2f}')
print(f'  Median Price: ${df["Modal Price"].median():.2f}')
print(f'\nQuantity Statistics:')
print(f'  Min Quantity: {df["Quantity"].min()}')
print(f'  Max Quantity: {df["Quantity"].max()}')
print(f'  Avg Quantity: {df["Quantity"].mean():.2f}')
```
